```python
import pandas as pd
import mysql.connector

conn = mysql.connector.connect(
    host='localhost',
    user='python_user',
    password='Python123!',
    database='accident_prediction'
)

df = pd.read_sql("SELECT * FROM accident_prediction_india", conn)
df.head(20)
```

```python
df.info()  
```

```python
df.describe() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='Accident Severity', y='Speed Limit (km/h)', data=df)
plt.title("Accident Severity vs Speed Limit")
plt.show()
```

```python
df.groupby('Year').size().plot(kind='bar', color='skyblue')
plt.title("Year-wise Accident Count")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
sns.scatterplot(x='Number of Vehicles Involved', y='Number of Casualties', data=df)
plt.title("Casualties vs Vehicles Involved")
plt.show()
```

```python
sns.histplot(df['Driver Age'], bins=15, kde=True, color="orange")
plt.title("Driver Age Distribution")
plt.show()
```

```python
sns.countplot(x='Accident Severity', data=df, palette="Set2")
plt.title("Accident Severity Distribution")
plt.show()
```

```python
df.groupby('Speed Limit (km/h)')['Accident Severity'].value_counts().unstack().plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Accident Severity by Speed Limit")
plt.show()
```

```python
df.to_excel("cleaned_accidents.xlsx", index=False)
print("Cleaned dataset exported to Excel!")
```
